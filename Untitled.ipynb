After checking for duplicates, null values and incorrect data type, it was discovered none of these anomalies were present but nonetheless an erreonous value was found. On the column age, we can observe a minimum value of -1, which must be an error as someone's age cannot be -1. 
We proceeded to then crosscheck any other value which could as well have an error of this nature for the Age column only given the other columns were judged to have coherent and consistent entries.

```python
#on the column age, we can observe a minimum value of -1, which must be an error as someone's age cannot be -1
#let us search for how many entries could have this error
error = df.query('Age <0')
error.count()['Age']
```

### Question 1 : What proprotion of patient with different types of illness show up for appoinment?

```python
#first we create a list of patient types
P_types = ['Hypertension','Diabetes','Alcoholism','Handicap']

#we create an array with the help of a for loop that will store:
#A) - the count for each disease
P_numbers = []

#B) - the count of patients that show up for appoinments, for each disease type
P_show_up = []  

#C) - the count of patients that have a scholarship ONLY, for each disease type
P_schol = []

#D) - the count of patients that received an sms ONLY and show up for appointments, for each disease type
P_sms_show_up = []

#E) - the count of patients that show up for appoinments and have a scholarship ONLY, for each disease type
P_schol_show_up = []

#F) - the count of patients that show up for appoinments that have a scholarship and received an sms, for each disease type
P_schol_sms = []
P_schol_sms_show_up = []


for p in P_types:
    illness = df[df[p]==1]
    ill_count = illness.count()[p]
    P_numbers.append(ill_count)
    print("Type of illness :",p)
    print("Number of Patients :",ill_count)
        
#what proportion of patients with each disease show up for appointments?
    ill_showup = illness.query('No_show=="No"')
    ill_show_count = ill_showup.count()[p]
    P_show_up.append(ill_show_count)
    print("Number of Patients who showed up for appoinments :",ill_show_count) 
    
#what proportion of patients with each disease show up for appointments after receiving an sms ONLY
    ill_show_sms = ill_showup.query('SMS_received==1').query('Scholarship==0')
    ill_show_sms_count = ill_show_sms.count()[p] 
    P_sms_show_up.append(ill_show_sms_count)
    print("Number of Patients who showed up after receiving an sms only:",ill_show_sms_count)
    
    
#what proportion of patients with each disease benefit from a scholarship without receiving an sms
    ill_schol = illness.query('Scholarship == 1').query('SMS_received==0')
    ill_schol_count = ill_schol.count()[p]
    P_schol.append(ill_schol_count)
    print("Number of patients having a scholarship: ",ill_schol_count)
    
    #what proportion now show up for appoinments
    ill_schol_show = ill_schol.query('No_show=="No"')
    ill_schol_show_count = ill_schol_show.count()[p]
    P_schol_show_up.append(ill_schol_show_count)
    print("Number of patients having a scholarship only and who showed up :",ill_schol_show_count)
    
#Now what proportion of patients benefit from a scholarship and receive sms
    schol_sms = illness.query('Scholarship == 1').query('SMS_received==1')
    schol_sms_count = schol_sms.count()[p]
    P_schol_sms.append(schol_sms_count)
    print("Number of patients having a scholarship and an sms:",schol_sms_count)
    
    #what proportion now show up for appointments
    schol_sms_showup = schol_sms.query('No_show == "No"')
    schol_sms_showup_count = schol_sms_showup.count()[p]
    P_schol_sms_show_up.append(schol_sms_showup_count)
    print("Number of patients having a scholarship and an sms who showed up :",schol_sms_showup_count)
    
P_numbers
P_show_up



```

```python
data = [P_numbers,P_show_up,P_schol,P_sms_show_up,P_schol_show_up,P_schol_sms,P_schol_sms_show_up]
X = np.arange(4)
plt.figure(figsize=(10,8))
#ax = fig.add_axes([0,0,1,1])
plt.bar(X + 0.00, data[0], color = 'red', width = 0.15)
plt.bar(X + 0.15, data[1], color = 'green', width = 0.15)
plt.bar(X + 0.30, data[2], color = 'blue', width = 0.15)
plt.bar(X + 0.45, data[3], color = 'yellow', width = 0.15)
plt.bar(X + 0.60, data[4], color = 'pink', width = 0.15)
plt.bar(X + 0.75, data[5], color = 'orange', width = 0.15)
plt.bar(X + 0.80, data[6], color = 'violet', width = 0.15)
plt.xticks(X,P_types)
plt.title("Distribution Of Patients grouped by Illness")
plt.legend(["Number Of Patients","Total Patients who showed up","showed up after receiving an sms only","having a scholarship only"
           ,"having a scholarship only and who showed up","having a scholarship and an sms","having a scholarship and an sms who showed up"])
```

### What proportion of patients with a scholarship show up for their appoinments?

```python
#first, we get the number of patients on a scholarship ONLY
schol = df.query('Scholarship==1').query('SMS_received==0')
schol_count = schol.count()['No_show']
schol_count
```

```python
#then we find the proportion of patients in this set that showed up for their appoinments
schol_show = schol[schol['No_show']=='No']
schol_show_count=schol_show.count()['No_show']
schol_show_count
```

```python
no_show = schol_count - schol_show_count
no_show
```

```python
#displaying the various proportions
labels=['Showed Up',"Didn't show up"]
plt.pie([schol_show_count,no_show],labels = labels, startangle=90, autopct='%.2f%%')
plt.axis('equal')
plt.title("Response to appoinments by patients with a scholarship")
```
